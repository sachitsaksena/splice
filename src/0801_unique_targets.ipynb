```python
import sys, os, fnmatch, datetime, subprocess, imp
import numpy as np
from collections import defaultdict
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
lib = pd.read_csv('061318_exonskipping_library.csv', index_col=0, usecols = ('Identifier number', 'Designed 61-bp target site (37i-24e, AG)', 'gRNA', 'Designed gRNA (NGG orientation)'))
```

```python
lib = lib.reset_index('Identifier number')
lib =  lib.rename(columns = {'Designed 61-bp target site (37i-24e, AG)':'target_site', 'Identifier number':'id'})
```

```python
lib.groupby(['id']).count()
```

```python
target_sites = lib.groupby('target_site').count().reset_index('target_site')
target_unique = target_sites.loc[target_sites.id == 1]

target_dict = pd.Series(target_unique.id.values, index = target_unique.target_site).to_dict()

lib['unique_target'] = lib.target_site.apply(lambda x: True if x in target_dict else False)
lib = lib.loc[lambda x: x.unique_target == True]
```

```python
lib.to_csv('unique_targets.csv', columns = ('id','target_site'), index = False)
```
